```python
import pandas as pd

# Update the file path to your downloaded file location
data = pd.read_csv('flight_data_2024.csv')

# Show first 5 rows to preview content
print(data.head())
```

```python
# Summary info: columns, data types, non-null counts
print(data.info())

# Basic statistics on numerical columns
print(data.describe())
```

```python
print("Columns in dataset:")
print(data.columns)
```

```python
missing = data.isnull().sum()
print("Missing values per column:")
print(missing[missing > 0])
```

```python
print("Total flights in dataset:", len(data))
print("Unique airlines:", data['op_unique_carrier'].nunique())
print("List of airlines:", data['op_unique_carrier'].unique())

import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(data['arr_delay'].dropna(), bins=50)
plt.title('Distribution of Arrival Delays (minutes)')
plt.xlabel('Arrival Delay (minutes)')
plt.show()

delayed = data[data['arr_delay'] > 15]
percent_delayed = (len(delayed) / len(data)) * 100
print(f"Percentage of flights delayed over 15 minutes: {percent_delayed:.2f}%")
```
